```python
%matplotlib inline

green_cow = cv.imread('res/green_cow.jpg')
green_cow_copy = np.copy(green_cow)

green_cow_copy = cvtColor(green_cow_copy,cv.COLOR_BGR2RGB)
plt.imshow(green_cow_copy)
```

```python
# Red Green Blue
lower_green= np.array([0, 200, 0])   # it is okay to have no red or blue but the lower value of green should be high
upper_green= np.array([170, 255, 205]) 
```

```python
#creating the mask over green area
mask= cv.inRange(green_cow,lower_green,upper_green)
```

```python
plt.imshow(mask,cmap='gray')
# mask ask each pixel if the color values fall in the range of upper and lower color threshold
# if it does fall the mask will allow it to display
# else it will block it out and turn pixel black
# 255 for white 0 for black
```

#### only showing cow

```python
mask_copy= np.copy(green_cow)

mask_copy[mask != 0 ] = [0,0,0]
plt.imshow(mask_copy)
```
