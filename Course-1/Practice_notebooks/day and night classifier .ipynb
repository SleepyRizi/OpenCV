```python
def avg_brightness(img):
    hsv_img=cvtColor(img,cv.COLOR_BGR2HSV)
    hsv_val=hsv_img[:,:,2]
    sum_brightness=np.sum(hsv_val)
    area= 600*1100.0 #pixels
    return sum_brightness/area
```

```python
print(avg_brightness(std_training_img[33][0]))
```
